```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load historic data and prediction data
historic_data = pd.read_csv('historic.csv')
prediction_data = pd.read_csv('prediction_input.csv')


```

```python
# Display basic information about the historic data
print("Historic Data Info:")
print(historic_data.info())

```

```python
# Display basic information about the prediction data
print("\nPrediction Data Info:")
print(prediction_data.info())

```

```python
# Explore the first few rows of each dataset
print("\nHistoric Data Head:")
print(historic_data.head())

print("\nPrediction Data Head:")
print(prediction_data.head())

```

```python
# Summary statistics of numerical features
print("\nSummary Statistics for Historic Data:")
print(historic_data.describe())

```

```python
# Distribution of the target variable (success_indicator) in historic data
plt.figure(figsize=(8, 6))
sns.countplot(x='success_indicator', data=historic_data)
plt.title('Distribution of Success Indicator in Historic Data')
plt.show()

# Visualize the correlation matrix for numerical features in historic data
plt.figure(figsize=(10, 8))
sns.heatmap(historic_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Historic Data')
plt.show()

```

```python
# Explore the distribution of categorical features
# For example, category, main_promotion, color

# Visualize the distribution of categories
plt.figure(figsize=(12, 8))
sns.countplot(x='category', data=historic_data)
plt.title('Distribution of Categories in Historic Data')
plt.xticks(rotation=45, ha='right')
plt.show()



```

```python
# Visualize the distribution of main_promotion
plt.figure(figsize=(12, 8))
sns.countplot(x='main_promotion', data=historic_data)
plt.title('Distribution of Main Promotions in Historic Data')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
# Visualize the distribution of colors
plt.figure(figsize=(12, 8))
sns.countplot(x='color', data=historic_data)
plt.title('Distribution of Colors in Historic Data')
plt.xticks(rotation=45, ha='right')
plt.show()
```
